```python
import os
```

```python
state_file_folder_path = '/app/hypatia/paper/satellite_networks_state/gen_data/sentinel2_isls_none_ground_stations_first_10_algorithm_free_one_only_gs_relays/dynamic_state_1000ms_for_3600s'
```

```python
state_filename_list = os.listdir(state_file_folder_path)
```

```python
for state_filename in state_filename_list:
    state_file_path = os.path.join(state_file_folder_path, state_filename)
    with open(state_file_path, 'r') as f:
        content = f.read()
        if content != "":
            print("state: " + state_filename + " with meaningful values:")
            print(content)
            print('\n')
```
